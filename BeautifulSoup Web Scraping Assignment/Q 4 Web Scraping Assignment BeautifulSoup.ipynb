```python
from bs4 import BeautifulSoup
import requests
```

```python
Presidt=requests.get("https://presidentofindia.nic.in/former-presidents.htm")
```

```python
Presidt
```

```python
Soup=BeautifulSoup(Presidt.content)
```

```python
Soup
```

```python
Name=[]
for i in Soup.find_all("h3"):
    Name.append(i.text)
Name
```

```python
Term=[]
for i in Soup.find_all("div",class_="presidentListing"):
     Term.append(i.find_all("p")[0].text.split(":")[1])
Term
```

```python
print(len(Name),len(Term))
```

```python
import pandas as pd
df= pd.DataFrame({"Name":Name,"Term":Term})
df
```
