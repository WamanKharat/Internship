```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet")
```

```python
page
```

```python
Soup=BeautifulSoup(page.content)
```

```python
Soup
```

```python
first_title=Soup.find('a',class_="restnt-name ellipsis")
first_title
```

```python
first_title.text
```

```python
loc=Soup.find("div",class_="restnt-loc ellipsis")
loc.text
```

```python
sta=Soup.find("span",class_="double-line-ellipsis")
sta
```

```python
sta.text.split("|")[0]
```

```python
title=[]
for i in Soup.find_all("a",class_="restnt-name ellipsis"):
    title.append(i.text)
```

```python
title
```

```python
Location=[]
for i in Soup.find_all("div",class_="restnt-loc ellipsis"):
    Location.append(i.text)
Location
```

```python
Price=[]
for i in Soup.find_all("span",class_="double-line-ellipsis"):
    Price.append(i.text)
Price
```

```python
image=[]
for i in Soup.find_all("img",class_="no-img"):
    image.append(i.get("data-src"))
image
```

```python
import pandas as pd
df= pd.DataFrame({"title":title,"Location":Location,"Price":Price,"image":image})
df
```
